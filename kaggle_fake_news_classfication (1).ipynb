```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Dense,LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size = 5000
```

```python

#one hot rep
```

```python
messages = x.copy()


```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
messages['title'][1]
```

```python

messages.reset_index(inplace=True)
```

```python
messages
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
import re
from nltk.corpus import stopwords

ps = PorterStemmer()
corpus = []
for i in range(0 , len(messages)):
  print(i)
  review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
  review = review.lower()
  review = review.split() # Split the review into a list of words

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
one_hot_rep = [one_hot(word,voc_size)for word in corpus]
one_hot_rep
```

```python
#perfom padding
sent_length = 20
embedded_docs = pad_sequences(one_hot_rep,padding = 'pre',maxlen = sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
#setup model to embedd and train
```

```python
model= Sequential()
```

```python
embedding_features = 40
model.add(Embedding(voc_size , embedding_features, input_length = sent_length))
model.add(LSTM(100))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
print(model.summary())
```

```python
import numpy as np
x_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x_final,y_final,test_size = 0.33, random_state = 42)
```

```python
model.fit(x_train,y_train,validation_data = (x_test,y_test),epochs = 20 , batch_size = 8)
```
